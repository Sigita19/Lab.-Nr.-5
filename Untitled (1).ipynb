```python
from sklearn.neighbors import DistanceMetric
from math import radians
import pandas as pd
import numpy as np

data = pd.read_csv('PS4_GamesSales1.csv', encoding="ISO-8859-1")

data['North America'] = np.radians(data['North America'])
data['Europe'] = np.radians(data['Europe'])
data['Japan'] = np.radians(data['Japan'])
data['Rest of World'] = np.radians(data['Rest of World'])
data['Global'] = np.radians(data['Global'])

data.head()
```

<h1>ChebyshevDistance<h1>

```python
dist = DistanceMetric.get_metric('chebyshev')
```

```python
data[['North America','Europe','Japan','Rest of World','Global']].to_numpy()
```

```python
 dist.pairwise(data[['Europe','Global']].to_numpy())
```

```python
 pd.DataFrame(dist.pairwise(data[['Europe','Global']].to_numpy()),  columns=data.Game.unique(), index=data.Game.unique())
```

```python
pd.DataFrame(dist.pairwise(data[['Europe','North America']].to_numpy()),  columns=data.Game.unique(), index=data.Game.unique())
```

```python
dist = DistanceMetric.get_metric('chebyshev')
dist.pairwise(data[['Europe','Japan']].to_numpy())
```
